```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt

```

```python
from numpy.random import *
```

```python
##line plot

x=np.arange(0,10)
y=np.arange(11,21)
print(x)
print(y)
```

```python
plt.plot(x,y,'go--',linewidth=2, markersize=12)
plt.title('Line plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.plot(x,y,'go--',linewidth=2, markersize=12)
plt.title("Line Plot")
plt.xlabel('x-axis')
plt.ylabel('y-label')
plt.show()
```
